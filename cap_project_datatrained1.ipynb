```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint

#Sno, Movie Name, Director Name, Duration, year, ratings, Metascore
#Movie Name, stars, votes, Genre, Gross collection, popularity, Certification
```

```python

movie_name=[]
dur=[]
year=[]
rating=[]
mtscore=[]
votes=[]
gross=[]
genre=[]
meta_sc=[]

```

```python
pages=np.arange(1,1750,50)
```

```python
for page in pages:
    page=requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup=BeautifulSoup(page.text,'html.parser')
    movie_data=soup.findAll('div',attrs={'class':'lister-item mode-advanced'})
    sleep(randint(2,8))
    for store in movie_data:
        name=store.h3.a.text
        movie_name.append(name)
        #----------------------------------------------------------------------------------
        year_of_rel=store.h3.find('span',class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
        year.append(year_of_rel)
        
        #----------------------------------------------------------------------------------
        runtime=store.p.find('span',class_="runtime").text.replace('min','')
        dur.append(runtime)
        
        #------------------------------------------------------------------------------------ 
        rate=store.find('div',class_="inline-block ratings-imdb-rating").text.replace('\n','')
        rating.append(rate)
        
        #------------------------------------------------------------------------------------
        meta=store.find('span',class_="metascore").text if store.find('span',class_="metascore") else " "
        meta_sc.append(meta)
        
        #----------------------------------------------------------------------------------
        genree=store.p.find('span',class_="genre").text.replace('\n','')
        genre.append(genree)
        
        #--------------------------------------------------------------------------------------
        value=store.find_all('span',attrs={'name':'nv'})
        vote=value[0].text.replace(',','')
        votes.append(vote)
        
        #-------------------------------------------------------------------------------------
        grosses=value[1].text.replace('$','').replace('M','') if len(value)>1 else " "
        gross.append(grosses)
     
    
```
